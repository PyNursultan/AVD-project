```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split

df['Cases'] = df['Cases'].fillna(df['Cases'].mean())  

train_data = df[df['Cases'].notnull()]
test_data = df[df['Cases'].isnull()]

train_data = pd.get_dummies(train_data, columns=['Items', 'Region'])
test_data = pd.get_dummies(test_data, columns=['Items', 'Region'])

print("Размер данных для обучения:", train_data.shape)
print("Размер данных для тестирования:", test_data.shape)

X_train = train_data.drop(columns=['Cases', 'Year'])  
y_train = train_data['Cases']  
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

X_test = test_data.drop(columns=['Cases', 'Year'])

X_test = X_test.reindex(columns=X_train.columns, fill_value=0)  

if X_test.shape[0] > 0:
     predictions = model.predict(X_test)

     df.loc[df['Cases'].isnull(), 'Cases'] = predictions

     print(df)
else:
    print("Тестовые данные пусты, проверьте обработку данных.")
```

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split


X = pd.get_dummies(df[['Year', 'Items', 'Region']], columns=['Items', 'Region'])
y = df['Cases']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=42)

model_rf = RandomForestRegressor(n_estimators=100, random_state=42)

model_rf.fit(X_train, y_train)

future_data = pd.DataFrame(columns=['Year', 'Items', 'Region', 'Predicted_Cases'])

def predict_for_year(year, items, region, model_rf, X_train):
    encoder = pd.DataFrame({'Year': [year], 'Items': [items], 'Region': [region]})
    encoder = pd.get_dummies(encoder, columns=['Items', 'Region'])
    encoder = encoder.reindex(columns=X_train.columns, fill_value=0)
    prediction = model_rf.predict(encoder)  
    return prediction[0]

def retrain_model(X, y, model_rf):
    model_rf.fit(X, y)

def run_single_session(model_rf, X_train, future_data, df):
    current_year = int(input("\nПожалуйста, введите год для предсказания (например, 2025): "))
    items = input("Введите значение для 'Items': ")
    region = input("Введите значение для 'Region': ")
    prediction = predict_for_year(current_year, items, region, model_rf, X_train)
    new_data = pd.DataFrame({
        'Year': [current_year],
        'Items': [items],
        'Region': [region],
        'Predicted_Cases': [prediction]
    })
    print(f"Предсказание для {current_year}: {prediction}")
    
    df = pd.concat([df, new_data], ignore_index=True)  

    X_train_new = pd.get_dummies(df[['Year', 'Items', 'Region']], columns=['Items', 'Region'])
    X_train_new = X_train_new.reindex(columns=X_train.columns, fill_value=0)  
    y_train_new = df['Cases'] 

    if X_train_new.isnull().sum().sum() > 0 or y_train_new.isnull().sum() > 0:
        X_train_new = X_train_new.fillna(0)  
        y_train_new = y_train_new.fillna(0)  

    retrain_model(X_train_new, y_train_new, model_rf)
    
    return model_rf, df

model_rf, df = run_single_session(model_rf, X_train, future_data, df)

# Вы можете вызвать функцию снова для следующего ввода
# model_rf, df = run_single_session(model_rf, X_train, future_data, df)
```
