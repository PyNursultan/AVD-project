```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split

df['Cases'] = df['Cases'].fillna(df['Cases'].mean())  

train_data = df[df['Cases'].notnull()]
test_data = df[df['Cases'].isnull()]

train_data = pd.get_dummies(train_data, columns=['Items', 'Region'])
test_data = pd.get_dummies(test_data, columns=['Items', 'Region'])

print("Размер данных для обучения:", train_data.shape)
print("Размер данных для тестирования:", test_data.shape)

X_train = train_data.drop(columns=['Cases', 'Year'])  
y_train = train_data['Cases']  
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

X_test = test_data.drop(columns=['Cases', 'Year'])

X_test = X_test.reindex(columns=X_train.columns, fill_value=0)  

if X_test.shape[0] > 0:
     predictions = model.predict(X_test)

     df.loc[df['Cases'].isnull(), 'Cases'] = predictions

     print(df)
else:
    print("Тестовые данные пусты, проверьте обработку данных.")
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df['Items'] = encoder.fit_transform(df['Items'])
df['Region'] = encoder.fit_transform(df['Region'])
```

```python
from sklearn.model_selection import train_test_split as tts
X = df.copy()
y = X.pop('Cases')
```

```python
X_train, X_test, y_train, y_test = tts(X, y, test_size = 0.3)
```

```python
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
```

```python
model_rf = RandomForestRegressor()
```

```python
model_gb = GradientBoostingRegressor()
```

```python
model_rf.fit(X_train, y_train)
```

```python
model_rf.score(X_train, y_train)
```

```python
model_rf.score(X_test, y_test)
```

```python
import pandas as pd
import numpy as np

future_years = list(range(2025, 2031))

future_data = pd.DataFrame({
    'Year': future_years,
    'Items': ['Alcoholophilia and alcohol psychosis', 'Drug addiction', 'Toxicomania', 'Registered with diagnosis', 'Alcoholophilia and alcohol psychosis', 'Drug addiction'],
    'Region': ['kyrgyzstan', 'Osh oblast', 'Ysykkul oblast', 'Talas oblast', 'Naryn oblast', 'Chui oblast']
})

print(future_data)

encoder = pd.get_dummies(future_data, columns=['Items', 'Region'])

encoder = encoder.reindex(columns=X_train.columns, fill_value=0)

predictions_future = model_rf.predict(encoder)

future_data['Predicted_Cases'] = predictions_future

print(future_data[['Year', 'Predicted_Cases']])
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.plot(future_data['Year'], future_data['Predicted_Cases'], marker='o', color='b', label='Predicted Cases')

plt.title('Predicted Cases for 2025-2030')
plt.xlabel('Year')
plt.ylabel('Predicted Cases')
plt.xticks(future_years)  
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
