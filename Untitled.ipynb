```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split

df['Cases'] = df['Cases'].fillna(df['Cases'].mean())  

train_data = df[df['Cases'].notnull()]
test_data = df[df['Cases'].isnull()]

train_data = pd.get_dummies(train_data, columns=['Items', 'Region'])
test_data = pd.get_dummies(test_data, columns=['Items', 'Region'])

print("Размер данных для обучения:", train_data.shape)
print("Размер данных для тестирования:", test_data.shape)

X_train = train_data.drop(columns=['Cases', 'Year'])  
y_train = train_data['Cases']  
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

X_test = test_data.drop(columns=['Cases', 'Year'])

X_test = X_test.reindex(columns=X_train.columns, fill_value=0)  

if X_test.shape[0] > 0:
     predictions = model.predict(X_test)

     df.loc[df['Cases'].isnull(), 'Cases'] = predictions

     print(df)
else:
    print("Тестовые данные пусты, проверьте обработку данных.")
```
